```python
import numpy as np

K=9100
Premium_call = 212
Premium_put = 113

Interval = 500
ST = np.arange(K-Interval, K+Interval)
#ST
Profit_longcall = np.maximum(ST-K,0) - Premium_call
Profit_shortcall = - Profit_longcall
Profit_longput = np.maximum(K-ST,0) - Premium_put
Profit_shortput = - Profit_longput
```

```python
import matplotlib.pyplot as plt
plt.subplot(211)
plt.plot(ST,Profit_longcall,lw=2,label="$LongCall$")
plt.plot(ST,Profit_shortcall,label="$ShortCall$")
plt.ylabel("Profit")
plt.xlabel("ST")
plt.legend()

plt.subplot(212)
plt.plot(ST,Profit_longput,lw=2,label="$LongPut$")
plt.plot(ST,Profit_shortput,label="$ShortPut$")
plt.ylabel("Profit")
plt.xlabel("ST")
plt.legend()

plt.show()
##
```

```python
##
K_1 = 9000; K_2 = 9200; K_3 = 9400
Premium_C1 = 282
Premium_C2 = 148
Premium_C3 = 57
Interval = 800
ST = np.arange(K_2 - Interval, K_2 + Interval)

Profit_longC1 = np.maximum(ST-K_1,0) - Premium_C1
Profit_short2C2 = 2*(Premium_C2 - np.maximum(ST-K_2,0))
Profit_longC3 = np.maximum(ST-K_3,0) - Premium_C3
Portfolio = Profit_longC1 + Profit_short2C2 +Profit_longC3
```

```python
import matplotlib.pyplot as plt
plt.plot(ST,Profit_longC1,"--",label="$LongC1$")
plt.plot(ST,Profit_short2C2,"--",label="$Short2C2$")
plt.plot(ST,Profit_longC3,"--",label="$ShortC3$")
plt.plot(ST,Portfolio,"black",lw=3,label="$Butterfly Spread$")
plt.ylabel("Profit")
plt.xlabel("ST")
plt.title("Butterfly Spread")
plt.ylim(-500,500)
plt.legend()

plt.show()
```
